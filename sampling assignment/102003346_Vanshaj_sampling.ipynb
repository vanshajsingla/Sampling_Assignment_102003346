```python
import pandas as pd
df=pd.read_csv('Creditcard_data.csv')
df
```

```python
print(df['Class'].value_counts())
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
print(X)
print(Y)
```

BALANCING CLASSES USING OVER SAMPLING

```python
from collections import Counter    
from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler(random_state=42)
x_ros,y_ros=ros.fit_resample(X,Y)
print('Original dataset shape:', Counter(Y))
print('Resample dataset shape', Counter(y_ros))
newdf=x_ros
newdf['Class']=y_ros
newdf
```

Sampling Techniques:

(I)   Simple Random Sampling-S1

(II)  Stratified Sampling-S2

(III) Cluster Sampling-S3

(IV)  Systematic Sampling-S4

(V)   Multistage sampling-S5

1. Simple Random Sampling - S1

```python
# n= (Z^2*p*(1-p))/E^2
# Taking
# Z= 1.96 (i.e. 95% confidence)
# p=0.5
# E=0.05
sample_size=int((pow(1.96,2)*0.5*0.5)/(pow(0.05,2)))
print(sample_size)
random_sample = newdf.sample(n=sample_size, random_state=0)
```

MODELS:

(I)  Random forest-M1

(II) KNN-M2

(III)Logistic regression-M3

(IV) Decision tree-M4

(V)  Gradient Boost-M5

```python
from sklearn.model_selection import train_test_split
X_random=random_sample.iloc[:,:-1]
Y_random=random_sample.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(X_random,Y_random,train_size=0.7)
```

```python
## Random forest-M1
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)
y_pred=rfc.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
## KNN-M2
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()   
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
## Logistic regression-M3
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
## Decision tree-M4
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train,y_train)
y_pred=dtc.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
## Gradient Boost-M5
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(x_train,y_train)
y_pred=xgb.predict(x_test)
accuracy_score(y_test,y_pred)
```

5. MULTISTAGE SAMPLING - S5

```python
# Count the number of positive and negative samples
n_negative,n_positive = newdf['Class'].value_counts()

# Calculate the weights for each sample
weights = np.zeros(len(newdf))
weights[newdf['Class'] == 1] = 1 / n_positive
weights[newdf['Class'] == 0] = 1 / n_negative
weights = weights / np.sum(weights)

# Set the sample size
sample_size = 500

# Perform weighted sampling
sample_indices = np.random.choice(range(len(newdf)), size=sample_size, replace=False, p=weights)
multistageSample=newdf.iloc[sample_indices]
X_multistageSample=multistageSample.iloc[:,:-1]
Y_multistageSample=multistageSample.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(X_multistageSample,Y_multistageSample,train_size=0.7)
multistageSample
```

```python
## Random forest-M1
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)
y_pred=rfc.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
## KNN-M2
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()   
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
## Logistic regression-M3
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
## Decision tree-M4
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train,y_train)
y_pred=dtc.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
## Gradient Boost-M5
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(x_train,y_train)
y_pred=xgb.predict(x_test)
accuracy_score(y_test,y_pred)
```
